## Popular Prompting Technniques

Prompting is guiding AI models with instructions or examples to get desired outputs. Common techniques include:

* **Zero-shot**: Direct instructions without examples.
* **Few-shot**: Providing a few clear examples.
* **Chain-of-thought (CoT)**: Asking models to explain their reasoning step-by-step.
* **Decomposition**: Breaking complex tasks into simpler subtasks.
* **Ensembling**: Combining multiple outputs to improve accuracy.
* **Self-Criticism**: Asking the model to critique and improve its own output.

## Examples

### 1. Zero-Shot Prompting

* **Definition**: Direct instructions without providing examples.
* **Example**:

  ```
  Classify the sentiment of this sentence as Positive, Negative, or Neutral:
  "The movie was incredible!"
  ```

### 2. Few-Shot Prompting (In-Context Learning)

* **Definition**: Using a few examples within the prompt.
* **Example**:

  ```
  Classify the sentiment:
  "I loved the meal!"  Positive
  "The service was awful."  Negative
  "It's an average place."  Neutral

  "The movie was incredible!"  ?
  ```

### 3. Chain-of-Thought (CoT)

* **Definition**: Asking the model to reason step-by-step.
* **Example**:

  ```
  Solve this math problem step-by-step:
  "If Sarah has 10 apples, eats 3, then buys 5, how many does she have?"
  ```

- #### **Types of Chain-of-Thought**

  - **1. Zero-Shot CoT**

  - Simply instruct the model to reason step-by-step without examples.
  - Effective with advanced models like GPT-4, Gemini 1.5.

  - **Example**:

  ```
  Q: "A bookstore had 40 books. It sold 12 books on Monday, and got 5 new books on Tuesday. How many books are left?"
  A: "Let's think step by step:
  - Start: 40 books
  - Sold 12 on Monday: 40 - 12 = 28
  - Received 5 on Tuesday: 28 + 5 = 33
  Thus, the answer is 33."
  ```


  - **2. Few-Shot CoT**

   - Provide explicit examples demonstrating step-by-step reasoning before the actual question.
   - Guides the model to follow a specific reasoning pattern.

   - **Example**:

  ```
  Q: "John had 6 apples, ate 2, then bought 4 more. How many apples does he have?"
  A: "Let's solve step by step:
  Start with 6 apples.
  Ate 2: 6 - 2 = 4.
  Bought 4 more: 4 + 4 = 8.
  Answer: 8 apples."

  Q: "A bookstore had 40 books. It sold 12 on Monday and got 5 new ones on Tuesday. How many books remain?"
  A: "Let's solve step by step:
    Start: 40 books.
    Sold 12: 40 - 12 = 28 books.
    Got 5: 28 + 5 = 33 books.
  Final Answer: 33 books."
  ```

---



### 4. Ensembling (Self-Consistency)

* **Definition**: Using multiple reasoning paths and choosing the most common answer.
* **Example**:

  ```
  Prompt the model 3 times with varied phrasing:
  - "Sarah had 10 apples, lost 3, then got 5 more. Count her apples."
  - "Start with 10 apples, subtract 3, add 5. Total apples now?"
  - "If 3 of Sarahs 10 apples disappear and she gets 5 more, what's her total?"

  Choose the most frequent answer.
  ```

### 5. Self-Criticism (Self-Refine)

* **Definition**: Prompting the model to critique and improve its own answer.
* **Example**:

  ```
  Question: "What's the capital of France?"
  Initial Answer: "Marseille"
  Critique: "Marseille is a major city but not the capital. The capital is Paris."
  Improved Answer: "Paris"
  ```

  * Models generate multiple reasoning paths, evaluate each, and systematically select the best solution, potentially revolutionizing problem-solving tasks.

### Further Insights
While recent language models have the ability to take long contexts as input, relatively little is known about how well they use longer context. We analyze the performance of language models on two tasks that require identifying relevant information in their input contexts: multi-document question answering and key-value retrieval. We find that performance can degrade significantly when changing the position of relevant information, indicating that current language models do not robustly make use of information in long input contexts. In particular, we observe that performance is often highest when relevant information occurs at the beginning or end of the input context, and significantly degrades when models must access relevant information in the middle of long contexts, even for explicitly long-context model
